```python
from IPython.display import display, HTML

def display_scrollable_dataframe(df, max_height=400):
    html = df.to_html(classes="scroll-table")
    style = f"""
    <style>
    .scroll-table {{
        height: {max_height}px;
        overflow: auto;
        display: block;
    }}
    .scroll-table thead {{
        position: sticky;
        top: 0;
        background-color: white;
    }}
    </style>
    """
    display(HTML(style + html))

display_scrollable_dataframe(df)

import pandas as pd
pd.set_option('display.max_rows', None)  # Show all rows


names = []
ids = []
duration = []
artists_list = []

# Extract data from the track dictionary
for track in tracks['items']:
    ids.append(track['track']['id'])
    names.append(track['track']['name'])
    duration.append(track['track']['duration_ms'] / 1000)
    
    # Get artist names
    artists = [artist['name'] for artist in track['track']['artists']]
    artists_list.append(artists)

# Find the max number of artists in any track
max_artists = max(len(artists) for artists in artists_list)

# Create columns dynamically
columns = ['id', 'name', 'duration'] + [f'artist_{i+1}' for i in range(max_artists)]

# Fill DataFrame
data = []
for i in range(len(ids)):
    row = [ids[i], names[i], duration[i]] + artists_list[i] + [''] * (max_artists - len(artists_list[i]))
    data.append(row)

df = pd.DataFrame(data, columns=columns)
df
```
